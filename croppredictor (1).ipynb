```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
data['soil']= label_encoder.fit_transform(data['soil'])
data['soil'].unique()
```

```python
from lazypredict.Supervised import LazyClassifier
from sklearn.model_selection import train_test_split
```

```python
X=data[['N', 'P', 'K', 'temperature', 'humidity', 'ph',
       'rainfall', 'min_price', 'max_price', 'modal_price',
       'soil']]
y=data[['commodity']]
from imblearn.over_sampling import SMOTE
from collections import Counter
smote = SMOTE()
X_train_os,y_train_os=smote.fit_resample(np.array(X),np.array(y))
X_train, X_test, y_train, y_test = train_test_split(X_train_os, y_train_os,test_size = 0.2, random_state = 202)
clf = LazyClassifier(verbose=0,ignore_warnings=True,custom_metric=None)
models, predictions = clf.fit(X_train, X_test, y_train, y_test)
print(models)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_model=RandomForestClassifier(random_state=200,max_depth=2,n_estimators=300,min_samples_leaf=7)
rf_model.fit(X_train,y_train)
prediction_train=rf_model.predict(X_train)
prediction_test=rf_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_train,prediction_train)
```

```python
accuracy_score(y_test,prediction_test)
```

```python
from sklearn.metrics import classification_report
print("classification report")
print(classification_report(y_train,prediction_train))
```

```python
from sklearn.metrics import mean_squared_error
train_mse=mean_squared_error(y_train,prediction_train)
train_mse
```

```python
test_mse=mean_squared_error(y_test,prediction_test)
test_mse
```
